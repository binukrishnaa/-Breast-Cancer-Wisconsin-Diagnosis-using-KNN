```python
import numpy as np
# numpy is used for numerical ,scientififc calculation
import pandas as pd
# pandas is used for data visualization and manipulation in python
import matplotlib as plt
# provides varies plotting  tools for data visualzation
import seaborn as sns
```

```python
df=pd.read_csv('/content/breast-cancer-wisconsin-data.csv')
df
# reading the csv file
```

```python
df.info()
# info is used for understanding the information of the csv file
```

```python
d={'B':0,'M':1}
# converting the diagnosis value of M and B to a numerical value (Malignant=1,Benign=0)
df['diagnosis']=df['diagnosis'].map(d)
# mapping them diagnosis column
df
```

```python
df.drop('id',axis = 1)
# dropping the id column
print(df.shape)
# shape is used to get the no of rows and columns i.e 569 is rows
# and 32 columns
df
```

```python
sns.lmplot(x = 'radius_mean', y = 'texture_mean', hue = 'diagnosis', data = df)
```

```python
sns.lmplot(x='smoothness_mean',y='compactness_mean',data=df,hue='diagnosis')
```

```python
x = np.array(df.iloc[:,2:])
y=np.array(df['diagnosis'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```
